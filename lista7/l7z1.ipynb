```
%matplotlib inline 

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

tests = [
         {
          'desc': 'parabola, 1-10-1, sig',
          'train': (np.linspace(-1, 1, 26), np.linspace(-1, 1, 26) ** 2),
          'model': tf.keras.Sequential([
            tf.keras.layers.Dense(10, activation='sigmoid'),
            tf.keras.layers.Dense(1)
            ]),
          'predict': np.linspace(-1, 1, 101)
         },
         {
          'desc': 'parabola, 1-10-10-1, sig, sig',
          'train': (np.linspace(-1, 1, 26), np.linspace(-1, 1, 26) ** 2),
          'model': tf.keras.Sequential([
            tf.keras.layers.Dense(10, activation='sigmoid'),
            tf.keras.layers.Dense(10, activation='sigmoid'),
            tf.keras.layers.Dense(1)
            ]),
          'predict': np.linspace(-1, 1, 101)
         },
         {
          'desc': 'sinus, 1-10-1, tanh',
          'train': (np.linspace(0, 2, 21), np.sin((3 * np.pi / 2) * np.linspace(0, 2, 21))),
          'model': tf.keras.Sequential([
            tf.keras.layers.Dense(10, activation='tanh'),
            tf.keras.layers.Dense(1)
            ]),
          'predict': np.linspace(0, 2, 161)
         },
         {
          'desc': 'sinus, 1-10-10-1, tanh, tanh',
          'train': (np.linspace(0, 2, 21), np.sin((3 * np.pi / 2) * np.linspace(0, 2, 21))),
          'model': tf.keras.Sequential([
            tf.keras.layers.Dense(10, activation='tanh'),
            tf.keras.layers.Dense(10, activation='tanh'),
            tf.keras.layers.Dense(1)
            ]),
          'predict': np.linspace(0, 2, 161)
         }
]

for test in tests:
  x, y = test['train']

  plt.title(test['desc'])
  plt.scatter(x, y)

  test['model'].compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
  test['model'].fit(x, y, epochs=100, verbose=False)
  predicted = []
  for p_x in test['predict']:
    predicted.append(test['model'].predict([p_x]))
  
  plt.scatter(test['predict'], predicted)
  plt.show()
```
